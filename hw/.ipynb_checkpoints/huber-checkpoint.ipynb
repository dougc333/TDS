Change notation from y to y-hat and f(x) to y. 
$\cfrac{\partial L(theta, y)}{\partial (dependentvars)}=0$
<p>For x inside quadratic area</p>
$\cfrac{\partial(loss)}{\partial \gamma}=2x^2(\hat y-y)$
$\cfrac{\partial(loss)}{\partial \beta}=2x(\hat y -y)$
$\cfrac{\partial(loss)}{\partial \alpha}=2(\hat y-y)$
<p>for x outside quadratic area</p>
$\cfrac{\partial(loss)}{\partial \gamma}=+/- x^2$
<p></p>
$\cfrac{\partial(loss)}{\partial \beta} = +/-x$
<p></p>
$\cfrac{\partial(loss)}{\partial \alpha}=+/-1$

<h6>Squared Function</h6>

```python
#It doesnt make sense to use a quadratic to approximate a linear function,so change the dataset to 
#be quadratic
def square_function(gamma, beta, alpha,x):
    return gamma*x**2 + beta*x + alpha

beta_true = 2.0
alpha_true = 2.0
gamma_true = 10.0
delta = .5
n = 300
x = random(n)
y = square_function(gamma_true,beta_true,alpha_true,x)+0.2*randn(n) + 4*binomial(1, 0.03, size=n) + 4*binomial(1,0.01,size=n)
beta = -1.
alpha = 3.
gamma = 2.
learning_rate = 0.01
```

```python
# plot and verify we have outliers
fig = plt.figure(dpi=80,figsize=(7,7))
ax = fig.add_subplot(111)
ax.set_xlim((0,1))
ax.set_ylim((-2,15))
plt.scatter(x,y,color="red")
line, = ax.plot([],[],lw=2)
```

```python
def is_outlier(x,y):
    #print(x[i],y[i])
    if(0<x and x<0.2 and y>5):
        #print("1:",x[i],y[i])
        return True
    elif(0.2<x and x<0.4 and y>6):
        #print("2:",x[i],y[i])
        #plt.scatter(x[i],y[i],color="red")
        return True
    elif(0.4<x and x<0.6 and y>8):
        #print("3:",x[i],y[i])
        return True
    elif(0.6<x and x<0.8 and y>10.5):
        #print("4:",x[i],y[i])
        return True
    elif(0.8<x and x<1.0 and y>15):
        #print("5:",x[i],y[i])
        return True
    else:
        return False


            
fig = plt.figure(dpi=80,figsize=(7,7))

ax = fig.add_subplot(111)
ax.set_xlim((0,1))
ax.set_ylim((-2,15))
for i in range(len(x)):
    if is_outlier(x[i],y[i]):
        plt.scatter(x[i],y[i],color="red")
    else:
        plt.scatter(x[i],y[i],color="blue")
print(len(y))
```

```python
def update_squared():
    global gamma, alpha, beta
    y_predicted = square_function(gamma,beta,alpha,x)
    #messed up you have to iterate through each traning example
    for i in range(n):
        print("x:",x[i]," y[i]",y[i])
        if not is_outlier(x[i],y[i]):
            der_wrt_gamma = (2/n)*(x[i]**2)*(y_predicted-y[i])
            der_wrt_alpha = (2/n)*(y_predicted-y[i])
            der_wrt_beta = (x[i]/n)*der_wrt_alpha
        else:
            der_wrt_gamma = (1/n)*(x[i]**2)
            der_wrt_alpha = (1/n)*x[i]
            der_wrt_beta = (1/n)*1
        alpha = alpha - learning_rate*der_wrt_alpha
        beta = beta - learning_rate*der_wrt_beta
        gamma = gamma - learning_rate * der_wrt_gamma
    
def animate(i):
    x = np.linspace(0,1,100)
    y = square_function(gamma,beta,alpha,x)
    line.set_data(x,y)
    for i in range(20): 
        update_squared()
    return (line,)
```

```python
#what is outlier supposed to do? does it become linear vs. squared loss? 



for i in range(2): 
    update_squared()
```

```python
fig = plt.figure(dpi=80,figsize=(7,7))
ax = fig.add_subplot(111)
ax.set_xlim((0,1))
ax.set_ylim((-2,15))
plt.scatter(x,y)
line, = ax.plot([],[],lw=2)
```

```python
resulting_animation = animation.FuncAnimation(fig,animate,frames=np.arange(0,250),interval=100)
```
