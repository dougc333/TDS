```python
import json
import os
import cv2

#test all files are there. 
test = json.load(open("malaria/test.json","r"))
train = json.load(open("malaria/training.json","r"))
print("num test:",len(test))
print("num train:",len(train))
num_test=0
num_train=0
#os.path.join needs leading / removed
print(test[0]['image']['pathname'])
num_test_found = 0
num_train_found = 0
not_found = 0

for i in range(len(test)):
    path_part = test[i]['image']['pathname'][1:len(test[0]['image']['pathname'])]
    p = os.path.join(os.getcwd(),'malaria',path_part)
    #print(p)
    try:
        img = cv2.imread(p)
        #print(img.shape)
        num_test_found += 1
    except:
        not_found +=1
        
for i in range(len(train)):
    path_part = train[i]['image']['pathname'][1:len(train[0]['image']['pathname'])]
    p = os.path.join(os.getcwd(),'malaria',path_part)
    #print(p)
    try:
        img = cv2.imread(p)
        #print(img.shape)
        num_train_found += 1
    except:
        not_found +=1
        
print("num_test:",num_test_found, " num_train:",num_train_found)  
print(not_found)
```

```python
#test bounding boxes
path_part = train[0]['image']['pathname'][1:len(train[0]['image']['pathname'])]
p = os.path.join(os.getcwd(),'malaria',path_part)
img = cv2.imread(p,cv2.IMREAD_COLOR)
for i in range(len(train[0]['objects'])):
    #print(train[0]['objects'][i])
    min_r = train[0]['objects'][i]['bounding_box']['minimum']['r']
    min_c = train[0]['objects'][i]['bounding_box']['minimum']['c']
    max_r = train[0]['objects'][i]['bounding_box']['maximum']['r']
    max_c = train[0]['objects'][i]['bounding_box']['maximum']['c']
    cv2.rectangle(img,(min_c,min_r),(max_c,max_r),(255,0,0), 2)
small = cv2.resize(img, None, fx=0.5, fy=0.5)
cv2.imshow("image", small)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
